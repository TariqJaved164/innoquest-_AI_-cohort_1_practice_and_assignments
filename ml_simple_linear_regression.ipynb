# Machine learning
## 1- Simple linear regression
### Step-1 Importing datasets

```python
# loading datasets
import pandas as pd
df = pd.read_csv('/content/sample_data/ml_data_salary.csv')
df.head()
```

## Step-2 Spliting data into training data and test data

```python
x=df[['YearsExperience']]
y=df[['Salary']]
```

```python
x.head()
```

```python
y.head()
```

```python
# import sklearn library
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
# load dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# Step-3 Fit linear regression model

```python
#from ast import mod
model=LinearRegression()
model.fit(x_train,y_train)
```

# Step-4 Poltting

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train)
plt.plot(x_train,model.predict(x_train),color='green')
plt.title('Salary vs Experience(Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_train,model.predict(x_train),color='green')
plt.title('Salary vs Experience(Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

# Step-5 Testing or evaluating of model

```python
# model fitness
print('Score for testing data= ',model.score(x_test,y_test))
print('Score for training data= ',model.score(x_train,y_train))
```

## Step-6 Predicting for unknown values
